# Condition Statement

```python
# if statement
age = 18

if age>= 18:
    print("You are allowed to vote in the elections")
```

```python
# else
## to determine eligible voters

age = int(input("Enter your age: "))

if age < 18:
    print("You are a minor, not eligible to vote")
else:
    print("You are eligible for voting")
```

```python
## elif
## the elif statement allows you to check multiple conditions. It stands for "else if"

age = int(input("Enter age: "))

if age < 18:
    print("You are a minor")
elif age >35:
    print("You're an old person")
elif age <= 35:
    print("You're a youth")
```

```python
## Nested Conditional Statements
'''
You can place one or more if, elif, or else statements inside another if, elif, or else statement
'''
num = int(input("Enter a number: "))

if num < 0:
    print(f"{num} is a negative number")
    if num%2==0:
        print(f"{num} is an even number")
    else:
        print(f"{num} is an odd number")
else:
    print("The number is zero or positive")
```

```python
# to determine a leap year
## it has to be divisible by 4 and 400

i_year = int(input("enter the year: "))

if i_year%4==0:
    if i_year%100==0:
        if i_year%400==0:
            print(i_year,"is a leap year")
        else:
            print(i_year, "not a leap year")
    else:
        print(i_year, "is a leap year")
else:
    print(i_year, "is not a leap year")
```

```python
# Simple Calculator program
#perform differnt operation

num_1 = float(input("Enter first number: "))
num_2 = float(input("Enter second number: "))
operation = input("Enter +, -, /, *, or ** for\
                  addition, subtraction, division, muliplication\
                   or square respectively")
if operation == '+':
    add = num_1 + num_2
    print(f"Addition of {num_1} and {num_2} is {add}")
elif operation == '-':
    sub = num_1 - num_2
    print(f"Addition of {num_1} from {num_2} is {sub}")
elif operation == '/':
    if num_2 == 0:
        print("Undefine")
    else:
        div = num_1/num_2
        print(f"division of {num_1} by {num_2} is {div}")
elif operation == '*':
    multiply = num_1 * num_2
    print(f"Multiplication of {num_1} by {num_2} is {multiply}")
elif operation == '**':
    power=num_1**num_2
    print(f"Raising {num_1} by the power {num_2} is {power}")
else:
    print("Error: invalid operation")
```
