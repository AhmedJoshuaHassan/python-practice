# for loop

```python
# range starts from index 0
range(5)
```

```python
for i in range(5):
    print(i)
```

```python
for i in range(1,11):
    print(i)
```

```python
range(1,10,2)
```

```python
for i in range(1,12,2):
    print(i)
```

```python
## strings
name ="Ahmed"

for i in name:
    print(i)
```
