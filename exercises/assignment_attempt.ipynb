```python
# Your code here
x = float(input("Enter first number: "))
y = float(input("Enter second number: "))
z = float(input("Enter third number: "))

if x > y and x>z:
    print(f"{x} is the greatest number between {x}, {y} and {z}")
elif y>x and y>z:
    print(f"{y} is the greatest number between {x}, {y} and {z}")
elif z>x and z>y:
    print(f"{z} is the greatest number between {x}, {y} and {z}")
```

**Question 5:** Write a Python program to calculate the factorial of a number.

```python
# Your code here
def factorial(n):
    if n == 0:
        return 1
    else:
        return n*factorial(n-1)
num = int(input("Enter a number: "))
print(f"The factorial of {num} is {factorial(num)}")
```

**Question 8:** Write a Python program to convert Celsius to Fahrenheit.

```python
# Your code here
print("This program convert's Celsius to Fahrenheit and vice versa")

what_convert = input("Enter C for Celsius to Fahrenheit or F for Fahrenheit to celsius: ")

if what_convert == "F":

    c_temp = float(input("Enter Temperature in degree celsius: " ))

    f_temp = round(c_temp*(9/5)+32,2)

    print(f"The temperature {c_temp} degree celsius is {f_temp} Fahrenheit")
elif what_convert == "C":
    f_temp = float(input("Enter Temperature in degree Fahrenheit: " ))

    c_temp = round(((f_temp-32)*5)/9,2)

    print(f"The temperature {f_temp} degree Fahrenheit is {c_temp} celsius")
```

**Question 13:** Write a Python program to demonstrate logical operators: and, or, not.

```python
# Your code here
ahmed_age = int(input("Enter your age Ahmed: "))
zainab_age = int(input("Enter your age Zainab: "))

print(f"Zainab is {zainab_age} years and Ahmed is {ahmed_age}")

if ahmed_age or zainab_age>20:
    print("These are adults")
else:
    print("They are still children")

if ahmed_age and zainab_age>20:
    print("These are adults")
else:
    print("They are still children")

if  not zainab_age>20:
    print("These are adults")
else:
    print("They are still children")
```
