```python
"""
### Assignment 1: Simple if Statement

Write a program that asks the user to input a number and prints whether the number is positive.

"""

num = int(input("Enter a number: "))

if num > 0:
    print(f"{num} is a positive number")
else:
    print("This is a zero or negative number")
```

```python
'''
### Assignment 2: if-else Statement
Write a program that asks the user to input a number 
and prints whether the number is positive or negative.
'''

num = int(input("Enter a number: "))

if num > 0:
    print(f"{num} is a positive number")
elif num<0:
    print(num, "is a negative number")
else:
    print("This is a zero")
```

```python
'''
### Assignment 4: Nested if Statement

Write a program that asks the user 
to input a number and prints whether the number is positive and even, 
positive and odd, or negative.
'''
num = int(input("Enter a number: "))

if num%2==0:
    if num>0:
        print(f"{num} is a positive number and even")
    else:
        print(f"{num} is a negative number and even")
else:
    if num<0:
        print(f"{num} is a negative number and odd")
    else:
        print(f"{num} is a positive number and odd")
```

```python
'''
### Assignment 5: for Loop
Write a program that prints all the numbers
 from 1 to 10 using a for loop.
'''

for i in range(1,11):
    print(i)
```

```python
'''
Write a program that prints all the numbers from 
1 to 10 using a while loop.
'''

n = 10
count = 0

while count<n:
    count = count+1
    print(count)
```

```python
'''
Write a program that prints a 
5x5 grid of asterisks (*) using nested loops.
'''

for i in range(5):
    for j in range(5):
        print(i, end=" ")
    print()
```

```python
total = 0

while True:
    number = int(input("Enter number, enter 0 when done: "))
    print(number)
    if number==0:
        break
    total +=number
print(total,"Sum of entered numbers")
```

```python
while True:
    num = int(input("Enter a number: "))
    if num==0:
        break
    for i in range(2,num+1,2):
        print(i)
    break

```

```python
def factorial(n):
    if n==0:
        return 1
    else:
        return n*factorial(n-1)

num = int(input("Enter a number: "))
num_f= factorial(num)
print(f"{num_f} is the factorial of {num}")
```

```python
number = int(input("Enter a number: "))
factorial =1
i = 1

while i<= number:
    factorial *= i
    i +=1
print(f"The factorial of {number} is {factorial}.")
```

```python
sum = 0

while True:
    num = int(input("Enter numbers, enter 0 when done: "))
    print(num)
    if num==0:
        break
    sum = sum + num

print(sum, "sum of numbers entered")
```

```python
n = int(input("Enter the number of Fibonacci numbers to print: "))
a, b = 0, 1
count = 0
while count < n:
    print(a)
    a, b = b, a + b
    count += 1
```
